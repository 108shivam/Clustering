```python
sns['gender']=sns['gender'].fillna(value='Other')
```

```python
sns1=pd.get_dummies(sns, columns=["gender"])
```

```python
# Handling the missing value method 2
def Missing_inputation(x):
    x=x.fillna(x.mean())
    return x
sns1=sns1.apply(lambda x: Missing_inputation(x))
```

```python
# handling outliers Method 2
def outlier_capping(x):
    x=x.clip(upper=x.quantile(0.99))
    x=x.clip(lower=x.quantile(0.01))
    return x

sns2=sns1.apply(lambda x: outlier_capping(x))
sns2.describe(percentiles=[.9,.95,.96,.97,.98,.99]).T
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def mykmeans(data,k,gap,verbose=True,grp=True):
    wcss=[]
    cluster=[]
    for i in range(1,k+1,gap):
        cluster.append(i)
        km=KMeans(n_clusters=i)
        km_result=km.fit(data)
        intr=km_result.inertia_
        wcss.append(intr)
        if verbose:
            print(f'for cluster {i} ; WCSS={round(intr, 2)}')
    if grp:
        plt.plot(cluster,wcss,label='withincluster sum of square(wcss)')
        plt.xlabel('Clusters')
        plt.ylabel('WCSS')
        plt.legend()
        plt.show()
        return wcss,cluster
```

```python
mykmeans(data=sns2,k=10,gap=1,verbose=True,grp=True)
```
